# **Air Quality Index Prediction**

Importing necessary Libraries and the required dataset

```python
import numpy as np
import pandas as pd

df=pd.read_csv('/content/data_date.csv')
df
```

Checking for the datatypes

```python
df.dtypes
```

Checking for missing values if any

```python
df.isna().sum()
```

Changing the format of date to numerical value

```python
df['Date']=pd.to_datetime(df['Date'])
```

Again checking for the datatypes

```python
df.dtypes
```

Creating seperate columns for year,month,day

```python
df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month
df['day']=df['Date'].dt.day
```

Printing df

```python
df
```

Now, again checking for datatypes

```python
df.dtypes
```

Droping the Date column from the dataset

```python
df=df.drop(['Date'],axis=1)
```

Printing df

```python
df
```

Performing Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Country']=lb.fit_transform(df['Country'])
df['Status']=lb.fit_transform(df['Status'])
```

Checking for the datatypes after performing Encoding

```python
df.dtypes
```

Seperating 'x' and 'y' columns from the dataset

```python
x=df.drop(['Status'],axis=1).values
y=df['Status'].values
```

Performing train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

Performing Normalization Technique

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

Creating a model using Naive Bayes Algorithm

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

Printing accuracy score for the above created model

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
```

```python
print("Accuracy Score for the model is =",ac)
```
