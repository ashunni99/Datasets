finding the object columns of the training data

```python
obj_cols=[col for col in df.columns if df[col].dtypes == 'object']
obj_cols
```

Performing Label Encoding for the object columns of the training data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for col in obj_cols:
  df[col]=lb.fit_transform(df[col])
```

Again checking for the datatypes of the training data after performing Encoding

```python
df.dtypes
```

finding the object columns of the testing data

```python
obj1_col=[col1 for col1 in df1.columns if df1[col1].dtypes == 'object']
obj1_col
```

Performing Label Encoding for the object columns of the testing data

```python
for col1 in obj1_col:
  df1[col1]=lb.fit_transform(df1[col1])
```

Again checking for the datatypes of the testing data after performing Encoding

```python
df1.dtypes
```

Seperating Input and Output from the datasets given

```python
x_train=df.iloc[:,:-1].values
y_train=df.iloc[:,-1].values
x_test=df1.iloc[:,:-1].values
y_test=df1.iloc[:,-1].values
```

Performing Random forest Algorithm for the above dataset (model creation)

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

Checking for the Accuracy Score of the created model

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
print("Accuracy score for the model is = ",ac)
```
