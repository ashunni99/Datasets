# **Autism Prediction**

Importing necessary functions and data (train and test seperate)

```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/train.csv')
df1=pd.read_csv('/content/test.csv')
```

Printing training data

Checking for the data types of training data

Performing Label Encoding for the Training Data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['gender']=lb.fit_transform(df['gender'])
df['ethnicity']=lb.fit_transform(df['ethnicity'])
df['jaundice']=lb.fit_transform(df['jaundice'])
df['austim']=lb.fit_transform(df['austim'])
df['contry_of_res']=lb.fit_transform(df['contry_of_res'])
df['used_app_before']=lb.fit_transform(df['used_app_before'])
df['age_desc']=lb.fit_transform(df['age_desc'])
df['relation']=lb.fit_transform(df['relation'])
```

Performing label encoding for the testing data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df1['gender']=lb.fit_transform(df1['gender'])
df1['ethnicity']=lb.fit_transform(df1['ethnicity'])
df1['jaundice']=lb.fit_transform(df1['jaundice'])
df1['austim']=lb.fit_transform(df1['austim'])
df1['contry_of_res']=lb.fit_transform(df1['contry_of_res'])
df1['used_app_before']=lb.fit_transform(df1['used_app_before'])
df1['age_desc']=lb.fit_transform(df1['age_desc'])
df1['relation']=lb.fit_transform(df1['relation'])
```

Identifying and applying values for x_train,y_train,x_test

```python
x_train=df.drop(['Class/ASD'],axis=1).values
y_train=df['Class/ASD'].values
x_test=df1
```

Performing K nearest neighbor algorithm for finding y_test

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_test=knn.predict(x_test)
```
