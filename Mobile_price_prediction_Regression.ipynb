```python
x_train=df.drop(['Price_Range'],axis=1).values
y_train=df['Price_Range'].values
x_test=df1.drop(['Price_Range'],axis=1).values
y_test=df1['Price_Range'].values
```

```python
# from sklearn.model_selection import train_test_split

# x_train, x_test, y_train, y_test = train_test_split(x_train,y_train,test_size = 0.3, random_state = 43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
lr=KNeighborsClassifier(n_neighbors=5)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
df2=pd.DataFrame({'Actual value':y_test,'Predicted value':y_pred,'Difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
```

```python
print("ACCURACY SCORE IS = ",ac)
```
